### Задание 5. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
from google.colab import files
import io

# Загрузка файла пользователем
print("Пожалуйста, загрузите файл titanic.csv.")
uploaded = files.upload()

# Проверяем, что файл был загружен
for file_name in uploaded.keys():
    print(f"Файл '{file_name}' успешно загружен.")

# Чтение файла titanic.csv
file_name = list(uploaded.keys())[0]  # Берём имя первого загруженного файла
data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Распределение переменной Pclass
pclass_distribution = data['Pclass'].value_counts().sort_index().to_dict()
print("Распределение Pclass (социально-экономический статус):")
for cls, count in pclass_distribution.items():
    print(f"Класс {cls}: {count} пассажиров")

# Распределение Pclass для мужчин
pclass_men = data[data['Sex'] == 'male']['Pclass'].value_counts().sort_index().to_dict()
print("\nРаспределение Pclass среди мужчин:")
for cls, count in pclass_men.items():
    print(f"Класс {cls}: {count} пассажиров")

# Распределение Pclass для женщин
pclass_women = data[data['Sex'] == 'female']['Pclass'].value_counts().sort_index().to_dict()
print("\nРаспределение Pclass среди женщин:")
for cls, count in pclass_women.items():
    print(f"Класс {cls}: {count} пассажиров")

# Подсчёт количества мужчин 2-го класса
men_in_second_class = pclass_men.get(2, 0)
print(f"\nКоличество мужчин 2-го класса: {men_in_second_class}")
```

### Задание 7. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
from google.colab import files
import io

# Загрузка файла пользователем
print("Пожалуйста, загрузите файл titanic.csv.")
uploaded = files.upload()

# Проверяем, что файл был загружен
for file_name in uploaded.keys():
    print(f"Файл '{file_name}' успешно загружен.")

# Чтение файла titanic.csv
file_name = list(uploaded.keys())[0]  # Берём имя первого загруженного файла
data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Фильтрация данных
# Группа моложе 30 лет
young_group = data[data['Age'] < 30]
young_survival_rate = young_group['Survived'].mean()  # Среднее значение в столбце 'Survived' даёт долю выживших

# Группа старше 60 лет
elderly_group = data[data['Age'] > 60]
elderly_survival_rate = elderly_group['Survived'].mean()  # Аналогично считаем долю выживших

# Вывод результатов
print(f"Доля выживших среди людей моложе 30 лет: {round(young_survival_rate * 100, 2)}%")
print(f"Доля выживших среди людей старше 60 лет: {round(elderly_survival_rate * 100, 2)}%")

# Проверяем гипотезу
if young_survival_rate > elderly_survival_rate:
    print("Люди моложе 30 лет выживали чаще, чем люди старше 60 лет.")
else:
    print("Люди моложе 30 лет не выживали чаще, чем люди старше 60 лет.")
```

### Задание 9. Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
from google.colab import files
import io
import re

# Загрузка файла пользователем
print("Пожалуйста, загрузите файл titanic.csv.")
uploaded = files.upload()

# Проверяем, что файл был загружен
for file_name in uploaded.keys():
    print(f"Файл '{file_name}' успешно загружен.")

# Чтение файла titanic.csv
file_name = list(uploaded.keys())[0]  # Берём имя первого загруженного файла
data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Фильтрация мужских имен
male_names = data[data['Sex'] == 'male']['Name']

# Функция для извлечения имени из полного имени
def extract_first_name(full_name):
    # Ищем шаблон имени в скобках или после титула (например, Mr., Dr.)
    match = re.search(r'\((.*?)\)|(?:Mr\.|Master\.|Dr\.|Rev\.)\s+([A-Za-z\-]+)', full_name)
    if match:
        # Если имя в скобках, берём его; иначе берём вторую группу (имя после титула)
        return match.group(1) or match.group(2)
    return None

# Применяем функцию к каждому имени
first_names = male_names.apply(extract_first_name).dropna()

# Находим самое популярное имя
most_common_name = first_names.value_counts().idxmax()
most_common_count = first_names.value_counts().max()

# Вывод результата
print(f"Самое популярное имя среди мужчин: {most_common_name} (встречается {most_common_count} раз)")
```

## Задание 10. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

### В среднем мужчины 1-го класса старше 40 лет.
### В среднем женщины 1-го класса старше 40 лет.
### Мужчины всех классов в среднем старше женщин того же класса.
### В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса

```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
from google.colab import files
import io

# Загрузка файла пользователем
print("Пожалуйста, загрузите файл titanic.csv.")
uploaded = files.upload()

# Проверяем, что файл был загружен
for file_name in uploaded.keys():
    print(f"Файл '{file_name}' успешно загружен.")

# Чтение файла titanic.csv
file_name = list(uploaded.keys())[0]  # Берём имя первого загруженного файла
data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Удаляем строки с отсутствующим возрастом
data = data.dropna(subset=['Age'])

# Средний возраст мужчин и женщин по классам обслуживания
avg_age_by_class_gender = data.groupby(['Pclass', 'Sex'])['Age'].mean()

# Средний возраст по классам
avg_age_by_class = data.groupby('Pclass')['Age'].mean()

# Вывод результатов
print("Средний возраст мужчин и женщин по классам обслуживания:")
print(avg_age_by_class_gender)
print("\nСредний возраст пассажиров по классам:")
print(avg_age_by_class)

# Проверка утверждений
# 1. В среднем мужчины 1-го класса старше 40 лет.
men_avg_age_class_1 = avg_age_by_class_gender.loc[(1, 'male')]
print(f"\nМужчины 1-го класса старше 40 лет: {'Да' if men_avg_age_class_1 > 40 else 'Нет'}")

# 2. В среднем женщины 1-го класса старше 40 лет.
women_avg_age_class_1 = avg_age_by_class_gender.loc[(1, 'female')]
print(f"Женщины 1-го класса старше 40 лет: {'Да' if women_avg_age_class_1 > 40 else 'Нет'}")

# 3. Мужчины всех классов в среднем старше женщин того же класса.
gender_comparison = all(
    avg_age_by_class_gender.loc[(cls, 'male')] > avg_age_by_class_gender.loc[(cls, 'female')]
    for cls in [1, 2, 3]
)
print(f"Мужчины всех классов старше женщин того же класса: {'Да' if gender_comparison else 'Нет'}")

# 4. Люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса.
class_age_comparison = avg_age_by_class[1] > avg_age_by_class[2] > avg_age_by_class[3]
print(f"В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса: {'Да' if class_age_comparison else 'Нет'}")
```

### Задание 11. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
# Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
from google.colab import files
import io

# Загрузка файла пользователем
print("Пожалуйста, загрузите файл titanic.csv.")
uploaded = files.upload()

# Проверяем, что файл был загружен
for file_name in uploaded.keys():
    print(f"Файл '{file_name}' успешно загружен.")

# Чтение файла titanic.csv
file_name = list(uploaded.keys())[0]  # Берём имя первого загруженного файла
data = pd.read_csv(io.StringIO(uploaded[file_name].decode('utf-8')))

# Удаляем строки с отсутствующим возрастом
data = data.dropna(subset=['Age'])

# Средний возраст спасённых и погибших
avg_age_survived = data[data['Survived'] == 1]['Age'].mean()
avg_age_not_survived = data[data['Survived'] == 0]['Age'].mean()

# Вывод результатов
print(f"Средний возраст спасённых пассажиров: {round(avg_age_survived, 2)} лет")
print(f"Средний возраст погибших пассажиров: {round(avg_age_not_survived, 2)} лет")

# Проверка утверждения
if avg_age_not_survived > avg_age_survived:
    print("Средний возраст погибших выше, чем у спасённых: Да")
else:
    print("Средний возраст погибших выше, чем у спасённых: Нет")
```
