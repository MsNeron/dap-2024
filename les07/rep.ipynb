### Задание 4. Удалите не числовые признаки, которые по вашему мнению, не могут влиять на заначение выжил или погиб пассажир.

```python
titanic_df = titanic_df.drop(['Name', 'Ticket', 'Cabin'], axis=1)
titanic_df
```

### Задание 5. Вместо признака "Sex" (я надеюсь вы его не удалили 😂) Создайте два новых признака male и female которые содержат значения 0 или 1.

```python
# Создание новых признаков male и female
titanic_df['male'] = (titanic_df['Sex'] == 'male').astype(int)
titanic_df['female'] = (titanic_df['Sex'] == 'female').astype(int)

# Удаление исходного признака "Sex"
titanic_df = titanic_df.drop('Sex', axis=1)
titanic_df
```

### Задание 6. Удалите строки, которые содержать хотя бы одно null значение

```python
titanic_df = titanic_df.dropna()
titanic_df.isnull().sum()
```

## **Исследование данных**

### Задание 4. Нормализуйте значения признака "Fare"

```python
titanic_df.loc[:, 'Fare'] = (titanic_df['Fare'] - titanic_df['Fare'].min()) / (titanic_df['Fare'].max() - titanic_df['Fare'].min())
titanic_df
titanic_df.isnull().sum()
```

### Задание 5. Найдите разницу между средними значениями признака "Fare" для погибших и выживших пассажиров.

```python
mean_fare_survived = titanic_df[titanic_df['Survived'] == 1]['Fare'].mean()
mean_fare_not_survived = titanic_df[titanic_df['Survived'] == 0]['Fare'].mean()

# Вычисление разницы
fare_difference = mean_fare_survived - mean_fare_not_survived
print(f"Разница между средними значениями 'Fare' для выживших и погибших пассажиров: {fare_difference:.2f}")
```

### Задание 6. Простройте на одной оси координат гистограмы значений признака "Fare" для погибших и выживших пассажиров.

```python
# Гистограммы для погибших и выживших пассажиров
plt.hist(
    titanic_df[titanic_df['Survived'] == 0]['Fare'],
    bins=30, alpha=0.7, color='red', label='Погибшие'
)
plt.hist(
    titanic_df[titanic_df['Survived'] == 1]['Fare'],
    bins=30, alpha=0.7, color='green', label='Выжившие'
)

# Добавление подписей и легенды
plt.title("Распределение стоимости билетов ('Fare') для погибших и выживших")
plt.xlabel("Стоимость билета ('Fare')")
plt.ylabel("Количество пассажиров")
plt.legend()

# Показ графика
plt.show()
```

### Задание 7. Сформируйте набор признаков (X). Сформируйте вектор целевых значений (y).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
X = titanic_df[["Pclass", "SibSp", "Parch", "female", "male", "Fare", "Age"]]
y = titanic_df["Survived"]
titanic_df.isnull().sum()
```

## **Предсказательная модель**

### Задание 8. Разделите набор данных на два, одни для обучения модели другой для проверки. Тестовый набор должен содержать 25 процентов данных

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.isnull().sum()
```

### Задание 9. Выполните обучение модели.

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=4, algorithm='auto')

model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

## **Проверка модели**

### Задание 10. Выведите мартицу ошибок

```python
from sklearn.metrics import confusion_matrix
model_matrix = confusion_matrix(y_test, y_pred, labels=[0,1])
model_matrix_df = pd.DataFrame(model_matrix)
model_matrix_df
```

### Задание 11. Расчитайте accuracy

```python
from sklearn.metrics import accuracy_score
model_accuracy = accuracy_score(y_test, y_pred)
round(model_accuracy,2)
```
